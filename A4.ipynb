```python
import pandas as pd

# 载入数据
train_data = pd.read_csv('./datasets/train.csv')
test_data = pd.read_csv('./datasets/test.csv')

# 数据预览
print("Training Data Preview:")
print(train_data.head())
print("Test Data Preview:")
print(test_data.head())

# 查看数据的基本信息
print("Training Data Information:")
print(train_data.info())
print("Test Data Information:")
print(test_data.info())

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# 忽略特定的警告
warnings.filterwarnings("ignore", category=FutureWarning)

# 如果您的数据集是一个文件，使用 pd.read_csv 或 pd.read_excel 来读取数据
# train_data = pd.read_csv('path_to_your_file.csv')

# 去掉ID列
train_data = train_data.drop(columns=['id'])

# # 描述性统计分析
# print("Descriptive Statistics of Training Data:")
# print(train_data.describe())

# 可视化目标变量“FloodProbability”的分布
plt.figure(figsize=(10, 6))
sns.histplot(train_data['FloodProbability'], kde=True, bins=30)
plt.title('Distribution of FloodProbability in Training Data')
plt.xlabel('FloodProbability')
plt.ylabel('Frequency')
plt.show()
```

分布形态：从直方图可以看出，“FloodProbability”呈现双峰分布，说明数据中可能存在两个或多个不同的子群体。这种分布可能暗示了不同的环境或地理特征对洪水概率的不同影响。

数据集中趋势：数据似乎主要集中在0.45到0.55的区间内，这可能表明在大多数情况下，洪水的可能性是中等的。

异常值：图表显示在两端的数据较少，表明极低和极高的洪水概率相对较少。这对于模型训练时处理样本不均匀的问题非常重要。
当你查看直方图时，它是按照下面的方式工作的：

X轴：表示数据值的范围，例如在你的图中，它表示“FloodProbability”的不同值，从最小到最大。
Y轴：表示频率，即在数据集中落在特定区间内的数据点数量。这告诉我们在某个具体的“FloodProbability”区间内有多少观测值。
例子
假设你有一个包含100个数据点的数据集，你绘制一个直方图，将数据分为若干个等宽的区间（或“桶”）。如果“FloodProbability”值从0.3到0.4的区间内有20个数据点，那么Y轴上对应这个区间的条形高度就会表示20。这个高度（频率）告诉我们，在这个概率区间内有20个观测值。

为什么频率重要？
数据分布：频率帮助我们理解数据如何在各个值之间分布。例如，大多数数据是否集中在某个特定区间？是否有异常值或者不常见的值区间？
统计分析：在进行统计测试和数据分析时，了解频率是至关重要的，因为它影响我们对数据集特性的理解。
模型建立：在建立预测模型时，了解目标变量的分布可以帮助我们选择合适的模型和预处理步骤，比如是否需要对数据进行变换来满足模型的假设（如正态性）。

```python

# # 去掉ID列
# train_data = train_data.drop(columns=['FloodProbability'])
# 计算相关系数矩阵
corr_matrix = train_data.corr()

# 绘制相关性热力图
plt.figure(figsize=(18, 18))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of All Features')
plt.show()

# 特别关注与'FloodProbability'相关的特征
flood_corr = corr_matrix['FloodProbability'].sort_values(ascending=False)
print("Correlation of features with FloodProbability:")
print(flood_corr)
```

相关性分布：所有特征与“FloodProbability”之间的相关性都相对较低，最高的相关性是“DeterioratingInfrastructure”（约0.19），这表明没有单一特征对洪水概率有决定性的影响。
特征相关性：特征间相关性普遍较低，这意味着大多数特征提供了独立的信息，对于构建预测模型来说是有利的。


这个热力图显示了多个特征之间的相关性矩阵，其中包括各种可能影响“FloodProbability”的因素，如“MonsoonIntensity”，“Urbanization”，“Deforestation”等。以下是从这个热力图中可以解读出的几点关键信息：

1. 自相关
图中左上角的红色对角线表示每个变量与自身的完全相关性（相关系数为1），这是相关矩阵的典型表示。
2. 特征间的相关性
红色区域：图中较暗的红色块表示较高的正相关性。例如，MonsoonIntensity与FloodProbability之间以及其他一些环境因素之间存在一定程度的正相关。
蓝色区域：图中的蓝色区域表示较低的相关性，越深的蓝色表示相关性越低。从图中可以看出，很多变量间的相关性不强。
3. 洪水概率的相关性
最后一行显示了所有特征与“FloodProbability”的相关性。可以观察到，大部分特征与“FloodProbability”的相关性不是特别强，但有少数几个特征显示出较明显的相关性。这些特征可能对预测洪水概率较为重要。
4. 模式与关系
该图还可以帮助识别可能的模式或者特征组合，这些模式可能在初步分析中不明显。例如，如果两个环境变量（如“Deforestation”和“Urbanization”）表现出相似的相关性模式，可能表明它们在影响“FloodProbability”方面具有相似的作用机制。

```python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures

# 假设您选择了'DeterioratingInfrastructure' 和 'MonsoonIntensity' 这两个特征
features = train_data[['DeterioratingInfrastructure', 'MonsoonIntensity']]

# 创建多项式和交互特征
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(features)

# 将结果转换为DataFrame
# 使用 get_feature_names_out 获取特征名称（确保您使用的 scikit-learn 版本是 0.24 或更高）
poly_features_df = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(features.columns))

# 从多项式特征DataFrame中删除原始特征列
poly_features_df = poly_features_df.loc[:, ~poly_features_df.columns.isin(['DeterioratingInfrastructure', 'MonsoonIntensity'])]

# 查看新生成的特征
print(poly_features_df.head())

# 将新特征合并到原始数据集中
extended_train_data = pd.concat([train_data, poly_features_df], axis=1)

# 查看扩展后的数据集
print(extended_train_data.head())
```

```python
# 计算扩展数据集的相关性矩阵
extended_corr_matrix = extended_train_data.corr()

# 打印与'FloodProbability'相关性的特征
print("Correlation of features with FloodProbability:")
print(extended_corr_matrix['FloodProbability'].sort_values(ascending=False))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 确保列名是DataFrame中存在的列
features_to_plot = ['DeterioratingInfrastructure^2', 'DeterioratingInfrastructure MonsoonIntensity', 'MonsoonIntensity^2']

# 添加目标变量列
features_to_plot.append('FloodProbability')

# 使用pairplot对选定的特征进行可视化，增加height参数调整子图的尺寸
sns.pairplot(data=extended_train_data[features_to_plot], height=3)  # height值可以根据需要调整，这里设置为3英寸
plt.show()

```

1. 单变量分布
FloodProbability 的分布呈现出中心倾向明显，主要集中在0.4到0.6之间。这表明大多数情况下，洪水概率是中等的，但同时也存在较低和较高概率的值，需要进一步分析这些极端值的成因。
2. 双变量关系
DeterioratingInfrastructure^2 vs. FloodProbability：数据点没有显示出明显的线性或非线性关系，分布较为分散，但整体趋势可能表明高DeterioratingInfrastructure平方值与较高的洪水概率略有相关性。
DeterioratingInfrastructure MonsoonIntensity vs. FloodProbability：这个交互特征与FloodProbability之间显示了更明显的非线性关系，随着DeterioratingInfrastructure MonsoonIntensity的增加，FloodProbability的值似乎也有所上升。这表明这两个特征的相互作用对洪水概率可能有重要影响。
MonsoonIntensity^2 vs. FloodProbability：这个关系图显示了一定的正相关趋势，尤其是在MonsoonIntensity^2值较高时，FloodProbability的值也相对较高。
对模型训练的启示
特征选择：基于上述分析，特别是DeterioratingInfrastructure MonsoonIntensity和MonsoonIntensity^2这两个特征似乎与FloodProbability的关系较强，应考虑将它们包括在内作为模型训练的输入特征。
非线性模型：由于一些特征与FloodProbability之间存在非线性关系，建议使用能够捕捉此类关系的模型，如决策树、随机森林或梯度提升机（GBM）。
数据预处理：在模型训练前，考虑对特征进行标准化或归一化，以处理不同特征之间的尺度差异。同时，也要注意处理任何可能的异常值或极端值，因为这些值可能会影响模型的性能。

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 分箱散点图：DeterioratingInfrastructure MonsoonIntensity vs. FloodProbability
plt.figure(figsize=(10, 6))
sns.scatterplot(x='DeterioratingInfrastructure MonsoonIntensity', y='FloodProbability', data=extended_train_data)
plt.title('DeterioratingInfrastructure MonsoonIntensity vs. FloodProbability')
plt.show()


```

该图显示随着 DeterioratingInfrastructure MonsoonIntensity 的增加，FloodProbability 的值在一定范围内变化较大，但在高值区域似乎有上升的趋势。这表明可能存在非线性关系，且该特征的较高值可能与较高的洪水概率关联。

```python
# 分箱散点图：MonsoonIntensity^2 vs. FloodProbability
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MonsoonIntensity^2', y='FloodProbability', data=extended_train_data)
plt.title('MonsoonIntensity^2 vs. FloodProbability')
plt.show()
```

类似地，随着 MonsoonIntensity^2 的增加，FloodProbability 在不同的值域内显示出多个水平的集群分布，可能指示不同的风险层次。这种模式可能表明多个因素共同作用于洪水概率。

```python
import numpy as np

# 计算`MonsoonIntensity^2`不同分位数下的FloodProbability
quantiles = np.percentile(extended_train_data['MonsoonIntensity^2'], [25, 50, 75])
for q in quantiles:
    subset = extended_train_data[extended_train_data['MonsoonIntensity^2'] <= q]
    print(f'Quantile {q}: Mean FloodProbability = {subset["FloodProbability"].mean()}')
```

分位数结果显示了不同的MonsoonIntensity^2分位数下FloodProbability的平均值。这些值随分位数的增加略有增长，支持了上述散点图中观察到的趋势。

```python
# 计算条件平均值
mean_conditions = extended_train_data.groupby('DeterioratingInfrastructure MonsoonIntensity')['FloodProbability'].mean()
print(mean_conditions)
```

条件平均值显示了DeterioratingInfrastructure MonsoonIntensity不同值下FloodProbability的平均表现。这些数据进一步证实了随着交互特征值的增加，洪水概率趋向于增加，尤其是在特征值较高时。
